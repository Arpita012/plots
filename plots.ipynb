```python
import pandas as pd
import plotly.express as px
df = pd.read_csv(r"C:\Assignment\cleaned_dataset\data\00001.csv")
#df = pd.read_csv(r"C:\Assignment\cleaned_dataset\data\00008.csv")

df['Battery_impedance'] =df['Voltage_measured'] / df['Current_measured']
df['Re'] = df['Battery_impedance'] * 0.7  
df['Rct'] = df['Battery_impedance'] * 0.3  

df['Cycle'] = range(1, len(df) + 1)

fig1 = px.line(df, x='Cycle', y='Battery_impedance', 
               labels={'Battery_impedance': 'Resistance (Ohms)', 'Cycle': 'Cycle Number'},
               title='Battery Impedance over Charge/Discharge Cycles')
fig1.show()

fig2 = px.line(df, x='Cycle', y='Re', 
               labels={'Re': 'Resistance (Ohms)', 'Cycle': 'Cycle Number'},
               title='Estimated Electrolyte Resistance (Re) over Charge/Discharge Cycles')
fig2.show()

fig3 = px.line(df, x='Cycle', y='Rct', 
               labels={'Rct': 'Resistance (Ohms)', 'Cycle': 'Cycle Number'},
               title='Estimated Charge Transfer Resistance (Rct) over Charge/Discharge Cycles')
fig3.show()
```
